```python
import pandas as pd
df=pd.read_csv('data1.csv')
df.head()
df.dtypes
```

```python
df = df[df['ENDPOINT_ID']=='bac14af4097f43b2b90274bd652c1391']
# df.head()
print(df)
```

```python
df['_time'] = pd.to_datetime(df['_time'])
df.head()
df.dtypes
```

```python
from pycaret.anomaly import AnomalyExperiment
s = AnomalyExperiment()
s.setup(df, session_id = 123)
```

```python
# # functional API
# iforest = create_model('iforest')
# print(iforest)

# OOP API
iforest = s.create_model('iforest')
print(iforest)
```

```python
# functional API
# models()

# OOP API
s.models()
```

```python
# functional API
# plot_model(iforest, plot = 'tsne')

# OOP API
s.plot_model(iforest, plot = 'tsne')
```

```python
# functional API
# plot_model(iforest, plot = 'umap')

# OOP API
s.plot_model(iforest, plot = 'umap')
```

```python
# functional API
# result = assign_model(iforest)
# result.head()

# OOP API
result = s.assign_model(iforest)
result.head()
```

```python
# functional API
# predictions = predict_model(iforest, data = data)
# predictions.head()

# OOP API
predictions = s.predict_model(iforest, data = df)
predictions.head()
```
